```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

import time
import pandas as pd


opciones=Options()
opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.add_argument('--incognito')                  # incognito
opciones.headless=False                               # si True, no aperece la ventana (headless=no visible)
opciones.add_argument('--start-maximized')            # comienza maximizado
#opciones.add_argument('user-data-dir=selenium')       # mantiene las cookies
#opciones.add_extension('driver_folder/adblock.crx')   # adblocker
```

```python
url = "https://www.clicars.com/"

driver = "./chromedriver.exe"
driver = webdriver.Chrome(driver , options = opciones)
driver.get(url)

driver.implicitly_wait(2) # is like a time.sleep but specific of selenium
driver.find_element_by_css_selector("#onetrust-accept-btn-handler").click()
driver.implicitly_wait(2) # is like a time.sleep but specific of selenium
driver.find_element_by_css_selector("#searcher-purchase > div > div.desktop-search-boxes.hidden-xs.col-xs-12 > div > div.bottom-buttons.d-flex.justify-content-center > button").click()
driver.implicitly_wait(2) # is like a time.sleep but specific of selenium

coches = []
for i in range(1,1000,2) :
    driver.execute_script("window.scrollBy(0, 100);behavior: 'smooth'")
    driver.implicitly_wait(1) # is like a time.sleep but specific of selenium
    if i == 95:
        try: driver.find_element_by_css_selector("#vehicles-list-page > section.sale-list > div.container > div:nth-child(4) > div > div > button").click()
        except: print("Todavía no ha salido el boton")
        
    print(i)
    dicc = {"i":i}
    try:
        dicc["Fabr_Modelo"] = driver.find_element_by_css_selector(f"#vehicles_list > article:nth-child({i}) > a > div > div.car-card-data-container > div.car-card-data.left > span.maker.ellipsis").text
        dicc["Descripcion"] = driver.find_element_by_css_selector(f"#vehicles_list > article:nth-child({i}) > a > div > div.car-card-data-container > div.car-card-data.left > span.version.ellipsis > strong").text
        dicc["Kilometraje"] = driver.find_element_by_css_selector(f"#vehicles_list > article:nth-child({i}) > a > div > div.car-card-data-container > div.car-card-data.left > span:nth-child(3) > small").text
        dicc["Precio_caro"] = driver.find_element_by_css_selector(f"#vehicles_list > article:nth-child({i}) > a > div > div.car-card-data-container > div.car-card-data.right.pl-2.col-auto > span.price").text
        dicc["Precio_desc"] = driver.find_element_by_css_selector(f"#vehicles_list > article:nth-child({i}) > a > div > div.car-card-data-container > div.car-card-data.right.pl-2.col-auto > span.priceweb > strong").text
    except:
        continue
    coches.append(dicc)

coches = pd.DataFrame(coches)
coches.to_csv("data/coches.csv")
```

```python

coches = pd.DataFrame(coches)
coches.to_csv("data/coches.csv")
coches
```

```python
coches[['Year','Kms','Power(CV)','Transmission']] = coches["Kilometraje"].str.split("|" , expand=True)
coches.drop(['Kilometraje'], axis=1, inplace = True)
coches['Kms']         = coches['Kms'        ].str.replace('km','')
coches['Precio_caro'] = coches['Precio_caro'].str.replace('€' ,'')
coches['Precio_desc'] = coches['Precio_desc'].str.replace('€' ,'')
coches['Power(CV)']   = coches['Power(CV)'  ].str.replace('CV','')
coches
```
